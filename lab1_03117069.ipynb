## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
seed = random.seed(seedno)
#n=4x μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
n = 49
d = 4
p = .0

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n,d,p,seed) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n = 49
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M, seed=seed)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n = 49
R = 0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R, seed=seed)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

#### Κώδικας υλοποίησης

```python
def cc(G, plot=True):
    node_clusters = nx.clustering(G)

    clustering_values = []
    for node in node_clusters:
      clustering_values.append(node_clusters[node])
    if plot:
      plt.hist(clustering_values)
      plt.title("Clustering coefficients histogram")
      plt.xlabel("Clustering coefficient")
      plt.ylabel("Number of nodes")

    mean = np.mean(clustering_values)
    print("Mean clustering coeff: {:.3f}".format(mean))
    std = np.std(clustering_values)
    print("Standard deviation: {:.3f}".format(std))

    max_node = max(node_clusters, key=node_clusters.get)
    min_node = min(node_clusters, key=node_clusters.get)
    print("Node with maximum clustering coefficient: {} with value {:.3f}".format(max_node, node_clusters[max_node]))
    print("Node with minimum clustering coefficient: {} with value {:.3f}".format(min_node, node_clusters[min_node]))
```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

#### Κώδικας υλοποίησης

```python
def compute_path_lengths_connected(G):
    all_lengths_d = dict(nx.all_pairs_shortest_path_length(G))
    all_lengths = []
    
    for node1 in all_lengths_d:
      for node2 in all_lengths_d[node1]:
        if node1 != node2:
          all_lengths.append(all_lengths_d[node1][node2])
    
    diameter = nx.algorithms.distance_measures.diameter(G)
    radius = nx.radius(G)
    mean = np.mean(all_lengths) if all_lengths else 0
    std = np.std(all_lengths) if all_lengths else 0
          
    print("Mean path length is: {:.3f}".format(mean))
    print("Standard deviation of path lengths: {:.3f}".format(std))
    print("The diameter of the network is: {:.3f}".format(diameter))
    print("The network's radius is: {:.3f}".format(radius))

def compute_path_lengths(G):
    if nx.is_connected(G):
      compute_path_lengths_connected(G)
    else:
      print("Graph is unconnected, work on each component separately")
      S = [(G.subgraph(c).copy(), c) for c in nx.connected_components(G)]
      
      for i in range(len(S)):
        print("\nFor subgraph no. {}:".format(i))
        if len(S[i][1])>2:
          compute_path_lengths_connected(S[i][0])
        else:
          print("The graph is too small to work on it")
      
```

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

##### Κώδικας

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο. 

def find_nodes_connected(G):
    all_l_d = dict(nx.all_pairs_shortest_path_length(G))

    diameter = nx.diameter(G)
    diam_cons = dict()

    for n1 in all_l_d:
      for n2 in all_l_d[n1]:    
        if all_l_d[n1][n2] == diameter:
          try:
            diam_cons[n1] += 1
          except:
            diam_cons[n1] = 1
    node_with_most = max(diam_cons, key=diam_cons.get)
    print("The node with the most connections of length equal to the diameter of the topology is node", 
          node_with_most, "with", diam_cons[node_with_most], "connections")
    
def find_nodes(G):
    if nx.is_connected(G):
      find_nodes_connected(G)
    else:
      print("Graph is unconnected, work on each component separately")
      S = [G.subgraph(c).copy() for c in nx.connected_components(G)]
      
      for i in range(len(S)):
        print("\nFor subgraph no. {}:".format(i))
        find_nodes_connected(S[i])
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

##### Κώδικας

```python
def find_periphery_connected(G):
    per = set()
    eccs = nx.eccentricity(G)
    diam = nx.diameter(G)

    for n in eccs:
      if eccs[n] == diam:
        per.add(n)

    print("Periphery of graph:", per)

def find_periphery(G):
    if nx.is_connected(G):
      find_periphery_connected(G)
    else:
      print("Graph is unconnected, work on each component separately")
      S = [G.subgraph(c).copy() for c in nx.connected_components(G)]
      
      for i in range(len(S)):
        print("\nFor subgraph no. {}:".format(i))
        find_periphery_connected(S[i])
```

```python
def find_center_connected(G):
    cntr = set()
    eccs = nx.eccentricity(G)
    rad = nx.radius(G)

    for n in eccs:
      if eccs[n] == rad:
        cntr.add(n)

    print("Center of graph:", cntr)

def find_center(G):
    if nx.is_connected(G):
      find_center_connected(G)
    else:
      print("Graph is unconnected, work on each component separately")
      S = [G.subgraph(c).copy() for c in nx.connected_components(G)]
      
      for i in range(len(S)):
        print("\nFor subgraph no. {}:".format(i))
        find_center_connected(S[i])
```

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

#### Κώδικας υλοποίησης

```python
def find_katz_connected(G):
    l_max = max(nx.adjacency_spectrum(G))
    katz_dict = nx.katz_centrality(G, alpha=1/np.real(l_max)-0.01)

    katz_vals = []
    for n in katz_dict:
      katz_vals.append(katz_dict[n])

    plt.hist(katz_vals)
    plt.show()
    
    print("The mean of Katz centrality in the network is: {:.3f}".format(np.mean(katz_vals)))
    print("Standard deviation is: {:.3f}".format(np.std(katz_vals)))


def find_katz(G):
    if nx.is_connected(G):
      find_katz_connected(G)
    else:
      print("Graph is unconnected, work on each component separately")
      S = [(G.subgraph(c).copy(), len(c)) for c in nx.connected_components(G)]
      
      for i in range(len(S)):
        print("\nFor subgraph no. {}:".format(i))
        if S[i][1] > 1:
          find_katz_connected(S[i][0])
        else:
          print("The graph has too few nodes to extract meaningful data")
```

### Συνολική κεντρικότητα

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

#### Κώδικας υλοποίησης 

```python
def find_central_connected(G):
    deg_dict = dict(nx.degree(G))
    top_degrees = sorted(deg_dict, key=deg_dict.get, reverse=True)[:10]

    close_dict = dict(nx.closeness_centrality(G))
    top_close = sorted(close_dict, key=close_dict.get, reverse=True)[:10]

    betw_dict = dict(nx.betweenness_centrality(G))
    top_betw = sorted(betw_dict, key=betw_dict.get, reverse=True)[:10]

    l_max = max(nx.adjacency_spectrum(G))
    katz_dict = nx.katz_centrality(G, alpha=1/np.real(l_max)-0.01)
    top_katz = sorted(katz_dict, key=katz_dict.get, reverse=True)[:10]


    print("The most central nodes by degree are: {}".format(top_degrees))
    print("By closeness: {}".format(top_close))
    print("By betweenness: {}".format(top_betw))
    print("By Katz centrality: {}".format(top_katz))

    central = []
    for n in top_degrees:
      if n in top_close and n in top_betw and n in top_katz:
        central.append(n)
    s = "\nNodes {} are ".format(central) if len(central)>1 else "Node {} is ".format(central)
    print(s+"in the top {} central nodes by all metrics".format(10))


def find_central(G):
    if nx.is_connected(G):
      find_central_connected(G)
    else:
      print("Graph is unconnected, work on each component separately")
      S = [(G.subgraph(c).copy(), len(c)) for c in nx.connected_components(G)]
      
      for i in range(len(S)):
        print("\nFor subgraph no. {}:".format(i))
        if S[i][1] > 2:
          find_central_connected(S[i][0])
        else:
          print("The graph has too few nodes to extract meaningful data")
```

```python
def pagerank(G):
    pr_d = nx.pagerank(G)
    
    pr_vals = []
    for n in pr_d:
      pr_vals.append(pr_d[n])
    pr_vals.sort(reverse=True)

    plt.hist(pr_vals)
    plt.title("PageRank score nodes")
    plt.ylabel("# of nodes")
    plt.xlabel("PageRank score")
    plt.show()
    
    print("The maximum PageRank in the network is: {:.5f}".format(np.max(pr_vals)))
    print("While the mean is {:.5f} and the standard deviation is {:.5f}".format(np.mean(pr_vals), np.std(pr_vals)))

pagerank(G40)
```

```python
print("Connectivity RG Gilbert")
ps = np.arange(1,11,1)
ps = [p/100 for p in ps]
connectivity = []
parameters = []

for n in ns:
  for p in ps:
    connected = 0
    for i in range(iters):
      G = nx.gnp_random_graph(n, p)
      connected += 1 if nx.is_connected(G) else 0
    connectivity.append(connected/iters*100)
    parameters.append(str((n,p)))

# print(connectivity)
plt.figure(figsize=(22,5))
plt.plot(connectivity)
plt.xticks(range(len(parameters)), parameters)
plt.title("Connectivity percentage for changing parameters")
plt.ylabel("Connectivity %")
plt.xlabel("(n,p) combination")
plt.show()
```

#### Να σχολιαστούν τα αποτελέσματα.

```python
def find_cut(G):
    if nx.is_connected(G):
      cut_nodes = nx.minimum_node_cut(G)
      print("The minimum number of nodes that if removed the graph will be disconnected is {}".format(len(cut_nodes)))
    else:
      print("The graph is already unconnected")
```

## E) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
random.seed(seedno)
n=309
d=4
gps = np.arange(0,11,1)
gps = [gp/10 for gp in gps]
# cc, compute_path_lengths

for gp in gps:
  print("For gp={}:".format(gp))
  G = nx.watts_strogatz_graph(n, d, gp)
  
  all_paths = []
  all_paths_dict = dict(nx.all_pairs_shortest_path_length(G))
  for n1 in all_paths_dict:
    for n2 in all_paths_dict[n1]:
      all_paths.append(all_paths_dict[n1][n2])

  avg_cc = nx.average_clustering(G)
  
  print("Average path length is {:.2f} and average clustering coefficient is {:.2f}\n".format(np.mean(all_paths), avg_cc))
```

Για $g_p$=0 το παραγόμενο δίκτυο φαίνεται να έχει τοπολογία RGG, αφού έχουμε μεγάλο μέσο μήκος μονοπατιού και μεγάλη συντελεστή ομάδοποίησης.<br>
Για $g_p$=1 το παραγόμενο δίκτυο φαίνεται να είναι τυχαίο, αφού έχουμε πολύ μικρό συντελεστή ομαδοποίησης και μικρά μονοπάτια.<br>
Για τιμές $g_p$ μεταξύ 0.1 και 0.5 παρατηρούμε σχετικά μικρές τιμές μέσου μήκους μονοπατιών και μεσαίες τιμές ομαδοποίησης, γεγονός που δείχνει σε τοπολογία small world.

#### Ποιοι ήταν οι 5 κεντρικότεροι χαρακτήρες του πρώτου κύκλου επεισοδίων ως προς το βαθμό, την εγγύτητα και την ενδιαμεσικότητα ?
